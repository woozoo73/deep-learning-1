# 초기화

```python
import tensorflow as tf

zeros = tf.zeros([10])
random_uniform = tf.random_uniform([1, 200], -1, 1)
random_normal = tf.random_normal([1, 200])

sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)

print("zeros\n", sess.run(zeros))
print("random_uniform\n", sess.run(random_uniform))
print("random_noraml\n", sess.run(random_normal))
```

## tf.random_uniform()

```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.plot(sess.run(random_uniform), 'ro', alpha=0.1)
plt.show()
```

## tf.random_normal()

```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.plot(sess.run(random_normal), 'ro', alpha=0.1)
plt.show()
```
